```
data2 = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/datasets/jd_source_c_2_filtered_clean_drop_bank_030622.csv")
```

```
data2
```

```
to_drop = ['Unnamed: 0.1']

data2.drop(to_drop, inplace=True, axis=1)
data2.head()
print(data2.head())
```

```
data2
```

```
data2 =data2.fillna(method ="ffill")
```

```
data2.head(30)
```

```
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```
data2["Unnamed: 0"] = LabelEncoder().fit_transform(data2["Unnamed: 0"] )
```

```
data2.head(30)
```

```
data2.rename(columns={"Unnamed: 0":"sentence_id","job_desc_clean":"words","Role":"labels"}, inplace =True)
```

```
data2["labels"] = data2["labels"].str.upper()
```

```
X= data2[["sentence_id","words"]]
Y =data2["labels"]
```

```
x_train, x_test, y_train, y_test = train_test_split(X,Y, test_size =0.2)
```

```
#building up train data and test data
train_data = pd.DataFrame({"sentence_id":x_train["sentence_id"],"words":x_train["words"],"labels":y_train})
test_data = pd.DataFrame({"sentence_id":x_test["sentence_id"],"words":x_test["words"],"labels":y_test})
```

```
train_data
```

# Model Training


```
from simpletransformers.ner import NERModel,NERArgs
```

```
label = data2["labels"].unique().tolist()
label
```

```
args = NERArgs()
args.num_train_epochs = 20
args.learning_rate = 1e-4
args.overwrite_output_dir =True
args.train_batch_size = 32
args.eval_batch_size = 32
```

```
model = NERModel('bert', 'bert-base-cased',labels=label,args =args)
```

```
model.train_model(train_data,eval_data = test_data,acc=accuracy_score)
```

```
result, model_outputs, preds_list = model.eval_model(test_data)
```

```
result
```

```
prediction, model_output = model.predict(["bank"])
```
