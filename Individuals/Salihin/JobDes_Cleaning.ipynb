```python
import pandas as pd
import numpy as np

import nltk
import re
nltk.download('punkt')
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem import SnowballStemmer
from nltk.tokenize import word_tokenize

# Load the specific job data to clean
data = pd.read_csv("Job_Physiotherapist.csv")
df = pd.DataFrame(data)

df
```

```python
df['Job Title'].is_unique
True
```

```python
jt = df['Job Title']
physio = jt.str.contains('Physiotherapist')
physio[:5]
```

**Change all job title into same title**

```python
df['Job Title'] = np.where(physio, 'Physiotherapist',                                  
                                    jt.str.replace('/', ' '))

df2 = df
df2
```

Cleaning the empty spaces and removing symbols

```python
df2['Company'] = df2['Company'].str.replace("\n","").str.replace("\t","").str.replace("\s+"," ").str.strip()

df2.to_csv('Testing.csv', index=False)
df2
```

**Create new column to be edited**

```python
df2['Job Description_Edited'] = df2['Job Description']

df3 = df2

df3
```

```python
import string

#removing emoji
def remove_emoji(string):
    emoji_pattern = re.compile("["
                           u"U0001F600-U0001F64F"  # emoticons
                           u"U0001F300-U0001F5FF"  # symbols & pictographs
                           u"U0001F680-U0001F6FF"  # transport & map symbols
                          u"U0001F1E0-U0001F1FF"  # flags (iOS)
                           u"U00002702-U000027B0"
                           u"U000024C2-U0001F251"
                           "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', string)


df3['Job Description_Edited'] = df3['Job Description_Edited'].str.replace("\n"," ").str.replace("\t","")\
            .str.replace("\d+","")\
            .str.replace(r'[/•!@#$%^&*()~↠=+☑✉►:;,<>?✅★◆✔\|–-♠♦_''""]',"").str.replace("\s+"," ")\
            .str.strip().str.lower()


df3.to_csv('Testing2.csv', index=False)
df3
```
